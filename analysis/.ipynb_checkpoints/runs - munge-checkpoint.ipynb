```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
import datetime as dt
from datetime import datetime as datetime
```

# Load runs

```python
filedata = None
with open("../data/1. raw/Garmin Connect Log - 2016-03-06.csv", 'r', encoding=None) as file :
  filedata = file.read()

# some text fields contain cr/lf or cr charactes.
# 1. Preserver current record delimiters by replacing "\r\n" with "|" (including quotes)
filedata = filedata.replace('"\r\n"', '"|"')
# 2. Remove all others cr/lf and cr characters
filedata = filedata.replace('"\r\n"', ' ')
# 3. Replace "|" back to "\r\n"
filedata = filedata.replace('"|"', '"\r\n"')
# some funny character in my data - no clue how it got there
filedata = filedata.replace(chr(0xa0), '') 

# Write the file out again - ignore errors with short lines
with open("../data/9. temp/Garmin Connect Log - 2016-03-06 - no CR.csv", 'w') as file:
  file.write(filedata)
```

```python
rr = pd.read_csv("../data/9. temp/Garmin Connect Log - 2016-03-06 - no CR.csv", 
                 index_col=False, error_bad_lines=False, encoding="ascii")
len(rr)

# there are warnings abot mismatching number of fields - those are manual entries and we can ignore them
```

# Leave only relevant columns

```python
rr = rr[["Activity ID",
        "Activity Name",
        "Begin Timestamp",
        "End Timestamp",
        "Activity Parent",
        "Activity Type",
        "Event Type",
        "Activity Time Zone",
        "Max. Elevation (Raw)",
        "Average Moving Speed (Raw)",
        "Max. Heart Rate (bpm)",
        "Average Heart Rate (bpm)",
        "Max. Speed (Raw)",
        "Calories (Raw)",
        "Duration (Raw Seconds)",
        "Moving Duration (Raw Seconds)",
        "Average Speed (Raw)",
        "Distance (Raw)",
        "Min. Elevation (Raw)",
        "Elevation Gain (Raw)",
        "Elevation Loss (Raw)"]]
```

```python
rr
```

# Convert factors to values

```python
rr["Begin Timestamp"] = rr["Begin Timestamp"].apply(lambda x: datetime.strptime(x, "%a, %b %d, %Y %H:%M %p"))
rr["End Timestamp"] = rr["End Timestamp"].apply(lambda x: datetime.strptime(x, "%a, %b %d, %Y %H:%M %p"))
```

# Delete irrelevant rows

```python
# select only running
rr = rr[rr["Activity Type"] == 'Street Running']
len(rr)
```

```python
# select only running, between 3 and 4 miles to eliminate long runs as they require different effort
rr = rr[(rr["Distance (Raw)"] > 3) & (rr["Distance (Raw)"] < 4)]

# select only 2013 and 2014
rr = rr[rr["Begin Timestamp"] < datetime(2015, 1, 1)]

# remove partially recorded run, which has outlayer value for 'Average Moving Speed (Raw)'
rr = rr[rr["Activity ID"] != 398755998]
len(rr)
```

# Save the new dataframe

```python
rr.to_csv("../data/2. cleansed/runs.csv", encoding="ascii")
rr.to_pickle("../data/2. cleansed/runs.pkl")
```
