```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
import datetime as dt
from datetime import datetime as datetime
```

# Load weather data, which covers interestin period:
Years: 2013-2015

```python
weather_files = ["722247-54785-2013", "722247-54785-2014", "722247-54785-2015"]
```

```python
tp = pd.DataFrame()
for f in weather_files:
    tp_temp = pd.read_fwf("".join(["../data/1. raw/", f]), header=None, index_col=False, 
                          widths=[4, 6, 5, 4, 2, 2, 2, 2, 1, 6,7, 5, 5, 5, 4, 3, 1, 1, 4, 1, 5, 1, 1, 1, 6, 1, 1, 1, 5, 1, 5, 1, 5, 1])  
    tp = tp.append(tp_temp, ignore_index=True)
len(tp)
```

```python
tp.head()
```

# Select only relevant columns

```python
tp = tp.iloc[:,[1, 2, 3, 4, 5, 6, 7, 9, 10, 12, 15, 18, 28, 30, 32]]
```

```python
tp.columns = ["USAFID", "WBAN", "YR", "M","D", "HR", "MIN", "LAT", "LONG", "ELEV", "WIND_DIR", "WIND_SPD", "TEMP", "DEW_POINT", "ATM_PRES"]
```

```python
tp.head()
```

# Scale values

```python
tp.LAT = tp.LAT/1000
tp.LONG = tp.LONG/1000
tp.WIND_SPD = tp.WIND_SPD/10
tp.TEMP = tp.TEMP/10
tp.DEW_POINT = tp.DEW_POINT/10
tp.ATM_PRES = tp.ATM_PRES/10 
  
```
