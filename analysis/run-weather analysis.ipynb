```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
import datetime as dt
from datetime import datetime as datetime
from ggplot import *
%matplotlib inline
```

```python
rw = pd.read_pickle("../data/2. cleansed/runs-weather.pkl")
```

```python
rw.info()
```

```python
def calc_color(date):
    """Calculate rgb tupple (0..1)"""
    return ((date.year - 2012)/(2015-2012), 
                   (date.month - 1)/(12-1), 0)
```

```python
rw["Color"] = rw["Begin Timestamp"].apply(calc_color)
```

```python
ggplot(rw, aes(x='Temperature', y='Average Speed (Raw)')) + \
    geom_point() + \
    geom_smooth(method='lm') + \
    geom_point(aes(colour = 'Color')) + \
    ggtitle("Average speed (Mile/min) by temperature") + \
    guides(colour = guide_legend(override.aes = list(shape = 15)))
    
```

```python
ggplot(rw, aes(x='Begin Timestamp', y='Average Speed (Raw)')) + \
    geom_line() + \
    ggtitle("Average speed (Mile/min) over time")
```
